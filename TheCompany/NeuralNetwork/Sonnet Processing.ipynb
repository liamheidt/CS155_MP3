# This script reads in and pre-processes the data

```python
import numpy as np

# First read the data in how it currently is

file1 = open("./data/shakespeare.txt","r") 
Data = file1.readlines()
file1.close()
Data
```

```python
# Make everything lower
for k in range(len(Data)):
    Data[k] = Data[k].lower()
```

```python
# Remove leading spaces
for k in range(len(Data)):
    while Data[k][0] == ' ':
        d = list(Data[k])
        Data[k] = "".join(d[1:len(d)])
```

```python
#Remove trailing spaces - I don't think there are any - so be careful (I don't know if this works)
for k in range(len(Data)):
    while Data[k][-1] == ' ':
        d = list(Data[k])
        Data[k] = "".join(d[0:-1])
```

```python
removeLines=['','\n']
removeLines.extend(['%i\n'%i for i in np.arange(1,155)])
removeLines=np.array(removeLines)
(removeLines==Data[51]).any()
```

## Remove empty lines (i,e ones with just '\n' or '')

```python
remove_indx = []
for k in range(len(Data)):
    if (removeLines==Data[k]).any():
        remove_indx.append(k)
```

```python
Data2 = []
for e in range(len(Data)):
    if e not in remove_indx:
        Data2.append(Data[e])
Data = Data2
```

```python
file=open('data/AllSonnets.txt','w')
for line in Data:
    file.write(line)
file.close()
```
